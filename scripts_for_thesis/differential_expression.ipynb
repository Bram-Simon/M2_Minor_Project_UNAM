```python
import pandas as pd
from scipy.stats import f_oneway
import numpy as np

def calculate_BH_adjusted_pvalues(df, p_value_col='p-value', FDR=0.05):
    """Applies the Benjamini-Hochberg procedure to adjust p-values
    
    df     :       DataFrame
        DataFrame containing the results
    p_value_col     :    String
        The name of the column containing p-values
    FDR:    :      int
        The false discovery rate (default: 0.05)
    """
    # Sort by p-value
    df = df.sort_values(p_value_col).reset_index(drop=True)
    
    # Calculate ranks
    df['rank'] = df.index + 1
    
    # Calculate BH critical values
    df['BH_critical_value'] = (df['rank'] / len(df)) * FDR
    
    # Adjusted p-values
    df['adjusted_p-value'] = df[p_value_col] * len(df) / df['rank']
    
    # Ensure adjusted p-values do not exceed 1
    df['adjusted_p-value'] = df['adjusted_p-value'].clip(upper=1.0)
    
    return df

# Initialize a list to hold ANOVA results
anova_results = []

N_tests = 0

# Perform ANOVA for each gene across the tumor sample groups
for gene in df_mex_tpms_model['gene_name']:
    # Extract gene expression data for the current gene across all samples
    gene_expression = df_mex_tpms_model[df_mex_tpms_model['gene_name'] == gene].drop("gene_name", axis=1)
    
    # Group the expression data based on the defined sample groups
    group_1_data = gene_expression[group_1].values.flatten()
    group_2_data = gene_expression[group_2].values.flatten()
    group_3_data = gene_expression[group_3].values.flatten()
    
    # Perform ANOVA if all groups have data
    if len(group_1_data) > 1 and len(group_2_data) > 1 and len(group_3_data) > 1:
        f_stat, p_val = f_oneway(group_1_data, group_2_data, group_3_data)
        anova_results.append({'Gene': gene, 'F-statistic': f_stat, 'p-value': p_val})
        N_tests += 1
    else:
        anova_results.append({'Gene': gene, 'F-statistic': np.nan, 'p-value': np.nan})

print("We performed {} statistical tests".format(N_tests))    

# Convert results to a DataFrame
df_anova = pd.DataFrame(anova_results)

# Apply the Benjamini-Hochberg procedure
df_anova = calculate_BH_adjusted_pvalues(df_anova)

# Display the results with adjusted p-values
display(df_anova)

for gene in df_mex_tpms_model['gene_name']:
    # Extract gene expression data for the current gene across all samples
    gene_expression = df_mex_tpms_model[df_mex_tpms_model['gene_name'] == gene].drop("gene_name", axis=1)

#df_significant_only = df_anova.copy()

for adj_pval in df_anova['adjusted_p-value']:
    df_significant_only = df_anova[df_anova['adjusted_p-value']<=0.05]#.drop('adjusted_p-value', axis=1)
#df_significant_only['adjusted_p-value' >= 0.05].drop
display(df_significant_only)
```
