```python
import matplotlib.pyplot as plt
import os
import pandas as pd
```

```python
# Code Block 1

# Get the current working directory
current_working_directory = os.getcwd()

# Print the current working directory
print(f"Current working directory: {current_working_directory}")
```

```python
# Code Block 2

base_dir = ".."
input_dir_v1 = os.path.join(base_dir, "scripts/OUT_CoRa_ATFv1")
input_dir_v2 = os.path.join(base_dir, "scripts/OUT_CoRa_ATFv2")
output_dir = os.path.join(base_dir, "output")

filename_v1 = "OUT_ExSSs_ATFv1_ATF_mY_mY.txt"
filename_v2 = "OUT_ExSSs_ATFv2_ATF_mY_mY.txt"
path_CoRa_results_v1 = os.path.join(input_dir_v1, filename_v1)
path_CoRa_results_v2 = os.path.join(input_dir_v2, filename_v2)
#CoRa_results = txt.read(path_CoRa_results)
```

```python
# Code Block 3

# Read the text file into a pandas DataFrame
# Adjust the delimiter based on your file's structure. For example, use delimiter='\t' for tab-separated values.
CoRa_results_df_v1 = pd.read_csv(path_CoRa_results_v1, delimiter='\t')
CoRa_results_df_v2 = pd.read_csv(path_CoRa_results_v2, delimiter='\t')

# Display the DataFrame to check the results
display(CoRa_results_df_v2)
```

```python
# Code Block 4

df_CoRa_values_v1 = CoRa_results_df_v1.iloc[:, -1] # Get CoRa values (final row)
df_CoRa_values_v2 = CoRa_results_df_v2.iloc[:, -1] # Get CoRa values (final row)
display(df_CoRa_values_v2)
```

```python
# Code Block 5

df_parameter_values_v1 = CoRa_results_df_v1.iloc[:, 0]
df_parameter_values_v2 = CoRa_results_df_v2.iloc[:, 0]

display(df_parameter_values_v2)
```

```python
def get_CoRa_plot(parameter_values, CoRa_values, title, width = 12, height = 5, size_title = 36,
                  size_ax_label = 31, tick_size = 25, width_line = 9):
    """
    Plots the CoRa (Concentration Ratio) values against perturbations.

    Parameters
    ----------
    parameter_values    :           list or numpy array
        Values of perturbations (mu_Y values) used in the analysis
    CoRa_values         :           list or numpy array
        Calculated CoRa values corresponding to the perturbations
    title               :           string
        Title for the plot and filename
    output_dir          :           string
        Directory to save the output plot
    width               :           float, optional
        Width of the figure in inches (default is 12)
    height              :           float, optional
        Height of the figure in inches (default is 6)
    size_title          :           int, optional
        Font size for the title (default is 32)
    size_ax_label       :           int, optional
        Font size for the axis labels (default is 28)
    tick_size           :           int, optional
        Font size for the ticks (default is 22)
    width_line          :           int, optional
        Width of the plot line (default is 7)
    
    Returns
    -------
    None
    """
    print("entering CoRa plot function")
    # Plot results
    plt.figure(figsize=(width, height))
    plt.plot(parameter_values, CoRa_values, linewidth = width_line)
    plt.xlabel('Y Synthesis Rate (muY)', size = size_ax_label)
    plt.ylabel('CoRa Value', size = size_ax_label)
    plt.xticks(size = tick_size)
    plt.yticks(size = tick_size)
    plt.xscale('log')
    plt.title(title, size = size_title)
    plt.tight_layout()
    #plt.savefig(os.path.join(output_dir, "{}.png".format(title)))
    fig_name_to_save = os.path.join(output_dir, "{}".format(title))
    plt.savefig(os.path.join(output_dir, "{}.png".format(fig_name_to_save)))
    plt.show()
    plt.close()
```

```python
title = "ATFv1 (fig 2C of CoRa paper reproduced)"
param_dfs = [df_parameter_values_v1, df_parameter_values_v2]
CoRA_dfs = [df_CoRa_values_v1, df_CoRa_values_v2]
get_CoRa_plot(param_dfs, CoRA_dfs, title)
```
